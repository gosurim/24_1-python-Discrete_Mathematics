```python
def tromino_tile(board, top, left, size, missing_x, missing_y, current_id):
    if size == 2: 
        # Base case: 2x2 board, place one tromino
        for i in range(2):
            for j in range(2):
                if board[top + i][left + j] == 0:
                    board[top + i][left + j] = current_id
        return current_id + 1  # Increment ID for the next tromino
    
    # Find center
    #원점
    half = size // 2
    center_x, center_y = top + half, left + half
    
    # Place tromino in the center if needed
    #사분면 지정
    missing_quadrant = 0
    if missing_x < center_x:
        if missing_y < center_y:
            missing_quadrant = 1
        else:
            missing_quadrant = 2
    else:
        if missing_y < center_y:
            missing_quadrant = 3
        else:
            missing_quadrant = 4
            
    # Place the tromino to cover the center of the 4 quadrants
    if missing_quadrant != 1:
        board[center_x - 1][center_y - 1] = current_id
    if missing_quadrant != 2:
        board[center_x - 1][center_y] = current_id
    if missing_quadrant != 3:
        board[center_x][center_y - 1] = current_id
    if missing_quadrant != 4:
        board[center_x][center_y] = current_id
        
    current_id += 1  # Increment ID for the next tromino
    
    # Recur for each quadrant
    # Top-left
    new_missing_x, new_missing_y = (center_x - 1, center_y - 1) if missing_quadrant != 1 else (missing_x, missing_y)
    current_id = tromino_tile(board, top, left, half, new_missing_x, new_missing_y, current_id)
    
    # Top-right
    new_missing_x, new_missing_y = (center_x - 1, center_y) if missing_quadrant != 2 else (missing_x, missing_y)
    current_id = tromino_tile(board, top, left + half, half, new_missing_x, new_missing_y, current_id)
    
    # Bottom-left
    new_missing_x, new_missing_y = (center_x, center_y - 1) if missing_quadrant != 3 else (missing_x, missing_y)
    current_id = tromino_tile(board, top + half, left, half, new_missing_x, new_missing_y, current_id)
    
    # Bottom-right
    new_missing_x, new_missing_y = (center_x, center_y) if missing_quadrant != 4 else (missing_x, missing_y)
    current_id = tromino_tile(board, top + half, left + half, half, new_missing_x, new_missing_y, current_id)
    
    return current_id
            
# Initialize the board and run the tiling algorithm
size = 16                        
missing = (0, size)                                      #오른쪽 맨 위 좌표 지정
board = [[0 for _ in range(size)] for _ in range(size)]  #board size
board[0][15] = -1                                        #-1을 넣어줍니다!

current_id = 1  #시작점??
tromino_tile(board, 0, 0, size, missing[0], missing[1], current_id)

# Print the board
for row in board:
        print(' '.join(str(cell).rjust(2) for cell in row))
```
